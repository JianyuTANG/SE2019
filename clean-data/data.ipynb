```python
import pandas as pd
import numpy as np
import os 

#显示所有列
pd.set_option('display.max_columns', None)
#显示所有行
pd.set_option('display.max_rows', None)
```

```python
sheets = ["1期","2期","3期","4期","5期","6期","7期","8期","9期","10期","20期"]
df_list = []
```

```python
for i in range(11):
    df = pd.read_excel("data.xls",sheet_name=sheets[i],dtype = str)
    df_list.append(df[["期数","姓名","本科院系代码","现常驻城市","期内职务"]])  
```

```python
df_list[10]
```

```python
member_df = pd.concat(df_list).reset_index(drop=True)
member_df
```

```python
member_df.drop_duplicates(subset=["姓名"],keep='first',inplace=False).reset_index(drop=True)
```
