```python
import pandas as pd
import numpy as np
import os 
import requests,json

#显示所有列
pd.set_option('display.max_columns', None)
#显示所有行
pd.set_option('display.max_rows', None)
```

```python
sheets = ["1期","2期","3期","4期","5期","6期","7期","8期","9期","10期","20期"]
df_list = []
```

```python
for i in range(11):
    df = pd.read_excel("data.xls",sheet_name=sheets[i],dtype = str)
    df_list.append(df[["姓名","期数","本科院系代码","现常驻城市"]])  
```

```python
member_df = pd.concat(df_list).reset_index(drop=True)
member_df["sessionCode"] = "admin"
member_df.columns = ['name', 'num', 'department', 'city', 'sessionCode']
member_df
```

```python
member_list = np.array(member_df).tolist()
member_list
```

```python
url = 'http://127.0.0.1:8000/add_userinfo'
for item in member_list:
    dic = dict(zip(['name', 'num', 'department', 'city', 'sessionCode'],item))
    data_json = json.dumps(dic)
    r_json = requests.post(url,data_json)
    print(r_json)
```
